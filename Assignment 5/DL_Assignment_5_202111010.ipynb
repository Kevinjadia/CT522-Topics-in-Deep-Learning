##Build CNN model and apply classification on MNIST-fashion dataset 





pipeline of cnn:<br>
1. understand the input & output(input image size)<br>
2. Train image and test image must have same size. As our model was trained on particular image size
3. If images are grayscale add fourth dimension to it(grayscale:fourth dimension is 1, RGB: fourth dimension is 3).
4. Normalize the image pixels by dividing it with 255
5. build cnn model layers and train and test on it.


```python
import keras
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
import numpy as np
from tensorflow.keras.datasets import fashion_mnist
import matplotlib.pyplot as plt
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.optimizers import Adam 
```

```python
batch_size = 128
num_classes = 10
epochs = 15
class_names=['T-shirt/top','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle boot']

img_rows, img_cols = 28, 28

# the data, split between train and test sets
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
x_train.shape
```

```python
y_train
```

```python

print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')

# If images are grayscale add fourth dimension to it
x_train = x_train.reshape(60000,28,28,1)
x_test = x_test.reshape(10000,28,28,1)

# convert class vectors to binary class matrices
y_train = tf.keras.utils.to_categorical(y_train, num_classes)
y_test = tf.keras.utils.to_categorical(y_test, num_classes)
```

```python
x_train=x_train/255.0
x_test=x_test/255.0
```

```python
#plt.imshow(x_test[0].reshape(28,28))
```

```python
# Padding = same: same as input size image after convolution 
# padding = valid: No padding --reduced size picture 
# filters = no of filters 
# kernel_size = filter size

fashion_mnist_model = tf.keras.models.Sequential()
fashion_mnist_model.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,padding="same", activation="relu", input_shape=[28,28,1]))
fashion_mnist_model.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,padding="same", activation="relu"))
fashion_mnist_model.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2,padding='valid'))
fashion_mnist_model.add(tf.keras.layers.Conv2D(filters=64,kernel_size=3,padding="same", activation="relu"))
fashion_mnist_model.add(tf.keras.layers.Conv2D(filters=64,kernel_size=3,padding="same", activation="relu"))
fashion_mnist_model.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2,padding='valid'))
# ANN model
fashion_mnist_model.add(tf.keras.layers.Flatten())
fashion_mnist_model.add(tf.keras.layers.Dropout(0.5,noise_shape=None,seed=None))
fashion_mnist_model.add(tf.keras.layers.Dense(units=128,activation='relu'))
fashion_mnist_model.add(tf.keras.layers.Dense(units=10,activation='softmax'))
fashion_mnist_model.summary()
```

```python
fashion_mnist_model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer=Adam(learning_rate = 0.0001))
fashion_mnist_model.fit(x_train,y_train,epochs=15)
```

```python
test_loss, test_accuracy = fashion_mnist_model.evaluate(x_test, y_test)
```

```python
print("Test accuracy: {}".format(test_accuracy))
```

```python
import os
import pathlib
import pickle
import numpy as np
import pandas as pd
import tensorflow as tf
from PIL import Image, ImageOps
from IPython.display import display
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
from matplotlib import rcParams
import warnings
```

```python
# Used PIL library to process images 

#The numpy.ravel() functions returns contiguous flattened array(1D array with all the input-array elements and with the same type as it). A copy is made only if needed.
def process_image(img_path: str) -> np.array:
    img = Image.open(img_path)
    img = np.ravel(img) / 255.0  #numpy.ravel() == numpy.reshape(-1)
    return img
```

```python
def process_folder(folder: pathlib.PosixPath) -> pd.DataFrame:
    # We'll store the images here
    processed = []
    
    # For every image in the directory
    for img in folder.iterdir():
        # Ensure JPG
        if img.suffix == '.jpg':
            # Two images failed for whatever reason, so let's just ignore them
            try:
                processed.append(process_image(img_path=str(img)))
            except Exception as _:
                continue
           
    # Convert to pd.DataFrame
    classes = pd.DataFrame(processed)
    # Add a class column - dog or a cat
    classes['class'] = folder.parts[-1]
    
    return processed,classes
```

```python
# Training set
train_berry,classes1 = process_folder(folder=pathlib.Path.cwd().joinpath('/content/Linnaeus 5 32X32/train/berry'))
train_bird,classes2 = process_folder(folder=pathlib.Path.cwd().joinpath('/content/Linnaeus 5 32X32/train/bird'))
train_dog,classes3= process_folder(folder=pathlib.Path.cwd().joinpath('/content/Linnaeus 5 32X32/train/dog'))
train_flower,classes4= process_folder(folder=pathlib.Path.cwd().joinpath('/content/Linnaeus 5 32X32/train/flower'))
train_other,classes5= process_folder(folder=pathlib.Path.cwd().joinpath('/content/Linnaeus 5 32X32/train/other'))

train_set = [train_berry,train_bird,train_dog,train_flower,train_other]
```

```python
x_train=np.array(train_set).reshape(6000, 32, 32,3)
```

```python
all_class=pd.concat([classes1, classes2,classes3,classes4,classes5], axis=0)
y_train = all_class['class']
```

```python
y_train=y_train.factorize()[0]
y_train
```

```python
# Testing set
test_berry,classes1 = process_folder(folder=pathlib.Path.cwd().joinpath('/content/Linnaeus 5 32X32/test/berry'))
test_bird,classes2 = process_folder(folder=pathlib.Path.cwd().joinpath('/content/Linnaeus 5 32X32/test/bird'))
test_dog,classes3= process_folder(folder=pathlib.Path.cwd().joinpath('/content/Linnaeus 5 32X32/test/dog'))
test_flower,classes4= process_folder(folder=pathlib.Path.cwd().joinpath('/content/Linnaeus 5 32X32/test/flower'))
test_other,classes5= process_folder(folder=pathlib.Path.cwd().joinpath('/content/Linnaeus 5 32X32/test/other'))

test_set = [test_berry,test_bird,test_dog,test_flower,test_other]
```

```python
x_test=np.array(test_set).reshape(2000, 32, 32,3)
```

```python
all_class=pd.concat([classes1,classes2,classes3,classes4,classes5], axis=0)
y_test = all_class['class']
```

```python
y_test=y_test.factorize()[0]
y_test
```

```python
num_classes=5
y_train = tf.keras.utils.to_categorical(y_train, num_classes)
y_test = tf.keras.utils.to_categorical(y_test, num_classes)
```

```python
# Padding = same: same as input size image after convolution 
# padding = valid: No padding --reduced size picture 
# filters = no of filters 
# kernel_size = filter size

cnn_model = tf.keras.models.Sequential()
cnn_model.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,padding="same", activation="relu", input_shape=[32,32,3]))
cnn_model.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,padding="same", activation="relu"))
cnn_model.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2,padding='valid'))
cnn_model.add(tf.keras.layers.Conv2D(filters=64,kernel_size=3,padding="same", activation="relu"))
cnn_model.add(tf.keras.layers.Conv2D(filters=64,kernel_size=3,padding="same", activation="relu"))
cnn_model.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2,padding='valid'))
# ANN model
cnn_model.add(tf.keras.layers.Flatten())
cnn_model.add(tf.keras.layers.Dropout(0.5,noise_shape=None,seed=None))
cnn_model.add(tf.keras.layers.Dense(units=128,activation='relu'))
cnn_model.add(tf.keras.layers.Dense(units=5,activation='softmax'))
cnn_model.summary()
```

```python
cnn_model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer=Adam(learning_rate = 0.001))
cnn_model.fit(x_train,y_train,epochs=15)
```

```python
test_loss, test_accuracy = cnn_model.evaluate(x_test, y_test)
```

```python
print("Test accuracy: {}".format(test_accuracy))
```
